Les variables les plus correlé avec ***margin_low***, ***margin_up*** sont ***length*** et ***is_genuine*** nous allons donc les utiliser pour l'imputation par regression linéaure

```python
# fonction d'imputation
def application_iterativeimputer(df, variables):
    df_imputed = df[variables]
    imp_median = IterativeImputer(random_state=21, missing_values=np.nan)
    imp_median.fit(df_imputed)
    df_imputed = imp_median.transform(df_imputed)
    return df_imputed
```

```python
varaibles = ["length", "is_genuine","margin_up", "margin_low"]
```

```python
# entrainement et imputation à travers la méthode application_iterativeimputer
margin_low_imputed = application_iterativeimputer(df_billets, varaibles)
```

```python
margin_low_imputed[:, 3]
```

```python
margin_low_imputed[:, 3].max()
```

```python
df_billets['margin_low'] = np.round(margin_low_imputed[:,3],2)
```

### Partie III

```python
df_copie.shape
```

```python
# 1) 
X = df_copie.drop(columns='is_genuine')
y = df_copie['is_genuine']
```

```python
print(X.shape)
print(y.shape)
```

```python
# 3) Division des données
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 2) Normalisation des données
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 4) Entrainement d'un premier model de regression logistique
regression_logistique = LogisticRegression()
regression_logistique.fit(X_train_scaled, y_train)
```

```python
# 5) Mesure des performances de ce model
predictions = regression_logistique.predict(X_test_scaled)
print("Regression logistique: " + str(accuracy_score(predictions, y_test)))
```

```python
print(predictions)
```

### II Optimisation

```python
# 1) définition du pipeline
pipeline = Pipeline(
    steps=[
        ("scaler", StandardScaler()),
        ("model",  LogisticRegression()),
    ]
)
```

```python
# 2
parameters = {
    "model__C": [1.0, 1.35,  1.75],
    "model__solver": ["newton-cg", "lbfgs", "bilinear"], 
    'model__penalty': ["l1","l2", "elasticnet", "none"]
}
```

```python
# 3) 
#define gridSearh 
grid_search = GridSearchCV(pipeline, parameters, cv=5, n_jobs=-1, verbose=1 )
```

```python
grid_search.fit(X_train , y_train) 
```

```python
best_model = grid_search.best_estimator_
```

```python
# 4) Valeurs des meilleurs hyperparamètres choisies par l'optimiseur
best_model._get_params
```

```python
predictions = best_model.predict(X_test)
```

```python
# 5) Performance du model
target_names = ['class 0', 'class 1']
print(classification_report(y_test, predictions, target_names=target_names))
```

```python
confusion_matrix(y_test, predictions)
```

#### Interpretations:

- precision: sur tous les billets que le model prédisait etre des vrais, seulement 99% le sont vraiment. Et sur tous les billets qu'il prédisait etre des faux seulement 99% le sont vraiment
- recall: sur tous billets identifiés positifs, le model a réuissi à prédire que 99% d'entre eux. Et sur tous les billets identifiés négatifs, le model a réussi à prédire la quasi totalité.
- fscore: sa veleur est très très proche de 1 donc on peut dire que le model est meilleur

```python
def truncate(number, n):
    number = str(number)
    number = number.split('.')
    number[1] = number[1][:n]
    number = '.'.join(number)

    return float(number)
```
