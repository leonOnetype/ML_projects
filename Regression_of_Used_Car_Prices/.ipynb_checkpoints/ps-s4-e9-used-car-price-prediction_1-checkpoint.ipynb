#### a) Case of the feature **engine**

Engine is usualy in the form: HorsePower(HP), Litres(L), Engine Configuration(number of cylinders) and  fluel type. So we will extract them.

```python
train["engine"].sample(5)
```

```python
def extract_engine_hp(engine_description):
    hp_pattern = r'(\d+\.?\d*)HP'
    # Extract horsepower
    hp_match = re.search(hp_pattern, engine_description)
    horsepower = float(hp_match.group(1)) if hp_match else None
    return horsepower
    
def extract_engine_liter(engine_description):
    liter_pattern = r'(\d+\.?\d*) ?L'
    # Extract liters
    liter_match = re.search(liter_pattern, engine_description)
    liters = float(liter_match.group(1)) if liter_match else None
    return liters

def extract_engine_config(engine_description):
    config_pattern = r'(?:(V?)(\d+)\s*Cylinders?)|(?:(I)(\d+))|(?:(H)(\d+))|(?:(W)(\d+))|(?:(Straight ?)(\d+)\s*Cylinders?)|(?:(V)(\d+))'
    
    # Extraction de la configuration du moteur
    config_match = re.search(config_pattern, engine_description)
    
    configuration = None
    if config_match:
        if config_match.group(2):  # V configuration
            configuration = int(config_match.group(2))
        elif config_match.group(4):  # I configuration
            configuration = int(config_match.group(4))
        elif config_match.group(6):  # H configuration
            configuration = int(config_match.group(6))
        elif config_match.group(8):  # W configuration
            configuration = int(config_match.group(8))
        elif config_match.group(10):  # Straight configuration
            configuration = int(config_match.group(10))
        elif config_match.group(12):  # V configuration without "Cylinders"
            configuration = int(config_match.group(12))
    
    return configuration

def extract_engine_fuel(engine_description):
    fuel_pattern = r'Hybrid|Flex Fuel|Diesel|Plug-In|Gasoline Fuel|Electric Fuel|Flexible Fuel|Hydrogen Fuel|GDI DOHC|GDI SOHC|GDI OHV|MPFI DOHC|MPFI OHV|DDI OHV|PDI DOHC|PDI SOHC|\
                        SIDI DOHC|TFSI|SC DOHC|SC ULEV|TSI|GTDI|MPFI|SOHC|DOHC'
    # Extract fuel type
    fuel_match = re.search(fuel_pattern, engine_description)
    fuel_type = fuel_match.group(0) if fuel_match else None

    if fuel_type != None: 
        fuel_type = fuel_type.replace("Fuel", "").strip()    # Remove the word Fuel
    elif 'Electric' in engine_description :   # If 
        fuel_type = "Electric"
    return fuel_type

def extract_engine_turbo(engine_description):
    turbo_pattern = r'(Turbo)'
    
    # Extract if Turbo or not
    turbo_match = re.search(turbo_pattern, engine_description)
    turbo = True if turbo_match else False

    return turbo 
```

```python
# On the train set
train["engine_hp"] = train["engine"].apply(extract_engine_hp)
train["engine_liters"] = train["engine"].apply(extract_engine_liter)
train["engine_config"] = train["engine"].apply(extract_engine_config)
train["engine_fuel_type"] = train["engine"].apply(extract_engine_fuel)
train["engine_turbo"] = train["engine"].apply(extract_engine_turbo)

# On the test set
test["engine_hp"] = test["engine"].apply(extract_engine_hp)
test["engine_liters"] = test["engine"].apply(extract_engine_liter)
test["engine_config"] = test["engine"].apply(extract_engine_config)
test["engine_fuel_type"] = test["engine"].apply(extract_engine_fuel)
test["engine_turbo"] = test["engine"].apply(extract_engine_turbo)
```

#### b) Case of the feature **transmission**

```python
list(train["transmission"].sample(5))
```

transmission is usualy in the form: number od speed, transmission type and shift mode. So we will extract them.

```python
def extract_nb_of_speed(transmission):
    nb_of_speed_pattern = r'(?i)(\d+)\-?\s*Speeds?|Single'

    # Extract 
    nb_of_speed_match = re.search(nb_of_speed_pattern, transmission)
    if nb_of_speed_match:
        try:
            nb_of_speed = int(str(nb_of_speed_match.group(1)))
        except ValueError:
            nb_of_speed = 1
    else:
        nb_of_speed=None
        
    return nb_of_speed

def extract_shift_mode(transmission):
    shift_mode_pattern = r'(Dual Shift|Auto-Shift)'
    # Extract 
    shift_mode_match = re.search(shift_mode_pattern, transmission)
    shift_mode = f"{shift_mode_match.group(0)}" if shift_mode_match else "Normal"

    return shift_mode 

def extract_transmission_type(transmission):
    transmission_type_pattern = r'(DCT|CVT|A/T|Automatic|Transmission w/Dual|AT|Mt|Manual|M/T|Variable)'
    # Extract 
    transmission_type_match = re.search(transmission_type_pattern, transmission)
    transmission_type = f"{transmission_type_match.group(0)}" if transmission_type_match else None

    return transmission_type 
```

```python
# On the train set
train["nb_of_speed"] = train["transmission"].apply(extract_nb_of_speed)
train["shift_mode"] = train["transmission"].apply(extract_shift_mode)
train["transmission_type"] = train["transmission"].apply(extract_transmission_type)

# On the test set
test["nb_of_speed"] = test["transmission"].apply(extract_nb_of_speed)
test["shift_mode"] = test["transmission"].apply(extract_shift_mode)
test["transmission_type"] = test["transmission"].apply(extract_transmission_type)
```

```python
def plot_distribution_and_average_price(dataframe, feature_name, xticks_rotation = 45, figsize = (20, 8), legend=True):
    plt.figure(figsize=figsize)

    # Colors
    colors = sns.color_palette("husl", dataframe[feature_name].nunique())
    data1 = dataframe[feature_name].value_counts()
    data2 = dataframe.groupby(feature_name)["price"].mean().sort_values(ascending=False)
    
    # bar_plot
    plt.subplot(1, 2, 1)
    bars = plt.bar(data1.index, data1, color=colors)
    plt.title(f"Distribution of {feature_name}")
    plt.xlabel(f"{feature_name}")
    plt.ylabel("Count")
    plt.xticks(rotation=xticks_rotation)  
    # Add a legende
    if legend==True:
        plt.legend(bars, data1.index, title=f"feature_name", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.subplot(1, 2, 2)
    bars = plt.bar(data2.index, data2, color=colors)
    plt.title(f" Average Price by {feature_name}")
    plt.xlabel(f"{feature_name}")
    plt.ylabel('Average price')
    plt.xticks(rotation=xticks_rotation)
    # Add a legende
    if legend==True:
        plt.legend(bars, data2.index, title=f"feature_name", bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show();  
```

```python
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.model_selection import KFold
from sklearn.base import clone
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
indexes = None
scaler = StandardScaler()
```

```python

random_state = 42
n_splits = 10

def Train_ML(model, X, y, test):
    global indexes
    
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    test_preds = np.zeros(test.shape[0])
    fold_idx = 0
    val_rmse_list = []

    for train_index, val_index in kf.split(X):
        X_train, X_val = X.iloc[train_index], X.iloc[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]

        if fold_idx == 4:
            indexes = {"train":train_index, "test":val_index}
            
        # Normalization
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)
        test_normalized = scaler.transform(test)

        # Training
        model_clone = clone(model)
        model_clone.fit(X_train, y_train)

        # Prediction and evaluation on the validation data
        val_preds = model_clone.predict(X_val)
        val_rmse = np.sqrt(mean_squared_error(y_val, val_preds))
        val_rmse_list.append(val_rmse)

        # Display infos
        print(f"Fold {fold_idx + 1}:")
        print(f"Validation RMSE: {val_rmse:.4f}")

        # Prediction on the test 
        test_preds_fold = model_clone.predict(test)
        test_preds += test_preds_fold

        fold_idx += 1
    
    mean_test_preds = test_preds / n_splits
    mean_val_rmse = np.mean(val_rmse_list)
    
    print('\n')
    print('==========================================')
    print(f"Mean Validation RMSE: {mean_val_rmse:.4f}")
    print('==========================================')

    return mean_test_preds
```

```python
lgb_params = {'learning_rate': 0.02383227411627239,
               'max_depth': 28, 
               'reg_alpha': 2.9369905019882525, 
               'reg_lambda': 9.06232547856336, 
               'num_leaves': 19, 
               'subsample': 0.7240326161291791, 
               'colsample_bytree': 0.4495893532949966}

lgb = LGBMRegressor(**lgb_params,random_state=random_state,verbose=-1,n_estimators=200)

pred_prices = Train_ML(lgb, X, y, X_test)
```
