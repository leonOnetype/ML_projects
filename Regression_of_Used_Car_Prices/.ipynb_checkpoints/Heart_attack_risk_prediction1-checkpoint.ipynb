### Description des features de la dataset

- **Age** : L'age du patient

- **Sex** : Le sex du patient

- **ExerciseAngina**: Est ce que la patient ressent de la douleur thoracique lorsqu'il fait des exercices physiques (1 = yes; 0 = no)

- **ChestPainType** : Type de douleur thoracique

    - Value 1: typical angina (douleur typique d'une crise cardiaque)
    - Value 2: atypical angina (qui ne correspond pas à la description classique de la crise cardiaque)
    - Value 3: non-anginal pain (qui n'est pas lié à en angine de poitrine)
    - Value 4: asymptomatic (abscence de douleur)
- **RestingBP**  :pression artérielle au repos. C'est à dire la pression artérielle mésurée lorsque la personne est dans un éta de repos, sans stress ni activité physique. (Exprimer en  mm Hg)

- **Cholesterol** : cholestérol en mg/dl recupérer via un capteur d'IMC

- **FastingBS** : fasting blood sugar(gycémie à jeun) ) (1 = true si > 120 mg/dl; 0 = false sinon)

- **RestingECG** : resting electrocardiographic results(Electrocardiogramme au repos)

    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- **thalach** : maximum heart rate achieved(Fréquence cardiaque maximale atteinte pendant un exercice physique). S'exprime en bpm(battements par minute)

- **MaxHR**: Fréquence cardiaque maximale en bpm.

- **oldpeak**: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)

-  **HeartDisease** : 
    - 0= Faible risque de crise cardiaque(ou pas de risque)
    - 1= Risque élevé de crise cardiaque

# Partie II: Feature selection and preprocessing

```python
df_copie = df_cardiaque.copy()
```

### II- 1 Sélection des features

Pour la sélection des features(variables) à fournir aux futurs modèles, nous allons nous baser sur la technique de sélection des variables basée sur le test de dépendance. Comme notre target est de type catégorielle,
1. Pour les variables numériques continues on utiliser le test d'Anova
2. Pour les variables vatégorielles on va utiliser le test du chi2
3. Enfin nous allons vérifier s'il n'existe pas deux features qui sont fortement corrélées entre elles. Si c'est le cas nous allons conserver une seule.

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif
```

Nous allons dabord dans un premier associer à chaque catégorie de chaque feature une valeur décimale unique. Ceci afin de pouvoir realiser la matrice de corrélation linéaire sur tout le dataset.

```python
for col in categorical_cols:
    
    # Extraction des valeurs uniques 
    unique_values = df_copie[col].unique() 
   
    # Création du dictionnaire de valeurs avec sa valeur numérique assiciée
    swap_values_dict = dict(zip(unique_values, [i for i in range(len(unique_values))]))
    
    # 
    print(swap_values_dict)
```

```python
for col in categorical_cols:
    
    # Extraction des valeurs uniques 
    unique_values = df_copie[col].unique() 
   
    # Création du dictionnaire de valeurs avec sa valeur numérique assiciée
    swap_values_dict = dict(zip(unique_values, [i for i in range(len(unique_values))]))
    
    # on fait un map
    df_copie[col] = df_copie[col].map(swap_values_dict)
    
# Vérification 
df_copie.head()
```

```python
categorical_cols
```

#### 1.1 Sélection des features catégorielles:Test du chi2

```python
X_cat, y = df_copie[categorical_cols], df_copie["HeartDisease"]
chi2_score, p_values = chi2(X_cat, y )
```

```python
p_values <= 0.05
```

la P-value represente la probabilité d'obtenir un résultat au moins aussi extreme que celui observé, sous l'hypothèse nulle(Hypothèse Nulle: Il y'a pas de lien avec la target)

```python
# Convertion de categorical_cols en tableau numpy afin de pouvoir faire le boolean indexing
categorical_cols = np.array(categorical_cols)
cat_features_selected = categorical_cols[p_values <= 0.05]
print(f"Les variables catégorielles sélectionnées:  {cat_features_selected}")
```

#### 1.2 Sélection des features numériques :Test d'Anova

```python
X_num, y = df_copie[numerical_cols], df_copie["HeartDisease"]
```

```python
f_statistic, p_values = f_classif(X_num, y)
f_statistic, p_values 
```

```python
# Convertion de numerical_cols en tableau numpy afin de pouvoir faire le boolean indexing
numerical_cols = np.array(numerical_cols)
num_features_selected = numerical_cols[p_values <= 0.05]
print(f"Les variables numériques sélectionnées:  {num_features_selected}")
```

```python
# Liste définitive des features sélectionnées
features_selected = np.concatenate([cat_features_selected, num_features_selected])
print(f"Liste des features sélectionnées: {features_selected}")
```

```python
final_df = df_copie[features_selected]
final_df.head()
```

```python
sns.heatmap(final_df.corr(), annot=True)
```

<font color=green> Interprétations: </font>
Il n'existe pas deux features parmis les features sélectionnées qui sont fortement correlées. Si c'était le cas, on allait conserver une seule. 

Avant de diviser les données nous allons faire une transformation sur les features catégorielles de notre nouveau dataset à l'exception de **ST_Slope**. En effet tel qu'elles sont representées dans le dataset, elles sont considérées comme des variables qualitatives ordinales or il y'a pas une relation d'ordre entre les valeurs prisent par ces features. Par exemple, **M** n'est pas supérieure ou inférieure à **F** ou encore on ne peut pas par exemple dire que **ATA** < **NAP** ou le contraire. Par contre pour la feature **ST_Slope** il y'a un ordre de grandeur entre ses différentes valeurs. Nous devons donc appliquer un autre encodage sur ceux-ci, l'encodage OneHot ou encore Dummy encodage.

```python
columns = list(set(cat_features_selected) - set(["ST_Slope"]))
final_df = pd.get_dummies(final_df, columns=columns)

#  Ajout de la target
final_df["target"] = df_copie["HeartDisease"]
final_df.head()
```

# Partie III: Modélisation

### III-1  Modélisation 

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline

# Les métrics
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay, accuracy_score
```

#### Définition de quelques models de classification

```python
lr = LogisticRegression()
lsvc = LinearSVC()
svc = SVC(kernel="rbf", C=1.0)
rf = RandomForestClassifier(criterion='gini', max_depth=150, min_samples_split=2, n_estimators=100, random_state=42)
knn = KNeighborsClassifier(n_neighbors=50)
sgd = SGDClassifier()
```

#### Définition de'une fonction metric pour l'évaluation des models

```python
def metric(y_test,y_pred):
    score = round(accuracy_score(y_test,y_pred),2)*100 
    print(f"Score = {score}%")
    ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
    #print(classification_report(y_test, y_pred))
    plt.show()
    return score
```

#### Logistic Regression

```python
pipe_lr = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", lr)
])
pipe_lr.fit(X_train,y_train)
pred_lr = pipe_lr.predict(X_test)
score_lr = metric(y_test, pred_lr)
```

#### SVC

```python
pipe_svc = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", svc)
])
pipe_svc.fit(X_train,y_train)
pred_svc = pipe_svc.predict(X_test)
score_svc = metric(y_test, pred_svc)
```

#### Linear SVC

```python
pipe_lsvc = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", lsvc)
])
pipe_lsvc.fit(X_train,y_train)
pred_lsvc = pipe_lsvc.predict(X_test)
score_lsvc = metric(y_test, pred_lsvc)
```

### Random Forest

```python
pipe_rf = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", rf)
])
pipe_rf.fit(X_train,y_train)
pred_rf = pipe_rf.predict(X_test)
score_rf = metric(y_test, pred_rf)
```

### Voting Classifier

```python
pipe_vt = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", VotingClassifier(estimators=[('lr',lr),('svc',svc),('lsvc', lsvc)],voting='hard'))
])
pipe_vt.fit(X_train,y_train)
pred_vt = pipe_vt.predict(X_test)
score_vt = metric(y_test, pred_vt)
```

#### KNN classifier

```python
pipe_knn = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", knn)
])
pipe_knn.fit(X_train,y_train)
pred_knn = pipe_vt.predict(X_test)
score_knn = metric(y_test, pred_knn)
print(score_knn)
```

#### SGD Classifier

```python
pipe_sgd = Pipeline(steps=[
    ("preprocesing", scaler),
    ("model", sgd)
])
pipe_sgd.fit(X_train,y_train)
pred_sgd = pipe_vt.predict(X_test)
score_sgd = metric(y_test, pred_sgd)
```

```python
data = {'models': ['Logistic Regression','SVC','Linear SVC','Random Forest Classifier', 'Voting Classifier', 'knn Classifier', "SGD Classifier"],
        'Score': [score_lr,score_svc,score_lsvc, score_rf, score_vt, score_knn, score_sgd]}
df =pd.DataFrame(data)
df['Score'] =df['Score']
df.sort_values(by='Score',ascending=False)
```

<font color=green> **Interprétation:** </font>
- Le modèle SVC semble etre meilleurs model il à précision élévée. Construisons les courbes d'apprentissages afin de voire comment s'est comporté nos model durant l'apprentissage.

### III-2 Courbes d'apprentissage

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import  learning_curve
```

#### Le KNN

```python
N, train_score, val_score = learning_curve(pipe_rf, X_train, y_train, cv=6,
                                              scoring="accuracy", train_sizes=np.linspace(0.001, 1, 100))
plt.figure(figsize=(5, 4))
plt.plot(N, train_score.mean(axis=1), label='train score')
plt.plot(N, val_score.mean(axis=1), label='val score')
plt.legend()
plt.title("Courbe d'apprentissage")
```

```python
params_grid_knn = {"model__n_neighbors": [ i for i in range(5, 15)], 
                   "model__weights":["uniform", "distance"],
                   "model__algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute'],
                    "model__p":[1, 2]}
```

```python
# 3) 
#define gridSearh 
grid_search = GridSearchCV(pipe_knn, params_grid_knn, cv=5, n_jobs=-1, verbose=1 )

grid_search.fit(X_train, y_train)

print("Meilleurs hyperparametres: ", grid_search.best_params_)

accuracy = grid_search.best_estimator_.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

```python
best_model = grid_search.best_estimator_
```

```python
N, train_score, val_score = learning_curve(best_model, X_train, y_train, cv=6,
                                              scoring="accuracy", train_sizes=np.linspace(0.001, 1, 100))
plt.figure(figsize=(5, 4))
plt.plot(N, train_score.mean(axis=1), label='train score')
plt.plot(N, val_score.mean(axis=1), label='val score')
plt.legend()
plt.title("Courbe d'apprentissage")
```

#### SVC

```python
N, train_score, val_score = learning_curve(pipe_svc, X_train, y_train, cv=6,
                                              scoring="accuracy", train_sizes=np.linspace(0.001, 1, 100))
plt.figure(figsize=(5, 4))
plt.plot(N, train_score.mean(axis=1), label='train score')
plt.plot(N, val_score.mean(axis=1), label='val score')
plt.legend()
plt.title("Courbe d'apprentissage")
```
